```python
import pandas as pd

customer_behavior = pd.read_csv("/content/Amazon Customer Behavior Survey.csv")
customer_behavior
```

```python
pd.set_option("display.max_rows", None)
customer_behavior["Improvement_Areas"].value_counts()
```

```python
customer_behavior_categorized = customer_behavior.copy()
```

```python
import uuid

customer_behavior["customer_id"] = [str(uuid.uuid4()) for _ in range(len(customer_behavior))]


customer_behavior["category_split"] = customer_behavior["Purchase_Categories"].str.split(";")
# Flatten the lists and find unique values
unique_categories = set([category for sublist in customer_behavior["category_split"].dropna() for category in sublist])
print(unique_categories)

# Expand each category into its own row
df_exploded = customer_behavior_categorized.explode("category_split")
# Rename the column for clarity
df_exploded = df_exploded.rename(columns={"purchase_category_split": "individual_category"})
# Display the expanded DataFrame
display(df_exploded)

```

```python
grouped = df_exploded.groupby(["category_split", "Gender"])["customer_id"].nunique()
grouped
```
