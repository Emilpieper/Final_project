```python

import pandas as pd

amazon = pd.read_csv("/content/Amazon/amazon_products.csv")

# Get the first chunk
#first_chunk = next(chunks)

# Display the first chunk
#display(first_chunk.head())  # Show the first 5 rows of the first chunk

categories = pd.read_csv("/content/Amazon/amazon_categories.csv")
```

```python
keywords = r'Sustainable|Organic cotton|Organic_cotton|Organic-cotton|'\
           r'sustainable cotton|Rainfed Cotton|Rainfed_Cotton|Rainfed-Cotton|sustainable_cotton|sustainable-cotton|'\
           r'Recycled|Upcycled|Biodegradable|Compostable|Organic|'\
           r'Natural fibers|Natural-fibers|Natural_fibers|'\
           r'Zero waste|Zero-waste|Zero_waste|Plastic-free|Plastic_free|Plastic free|'\
           r'Bamboo|Chemical-free|Chemical_free|Chemical free|'\
           r'Sustainable packaging|Sustainable-packaging|Sustainable_packaging|'\
           r'Handmade|Hemp|Green living|Green-living|Green_living|'\
           r'Sustainable lifestyle|Sustainable-lifestyle|Sustainable_lifestyle|'\
           r'Eco-conscious|Eco_conscious|Eco conscious|'\
           r'Renewable materials|Renewable-materials|Renewable_materials|'\
           r'Plant-based|Plant_based|Plant based|Eco-friendly material|Eco_friendly_material|Eco friendly material|'\
           r'Non-toxic|Non_toxic|Non toxic|Certified Organic|Certified-Organic|Certified_Organic|'\
           r'USDA Organic|USDA-Organic|USDA_Organic|EPEAT|'\
           r'Carbon Neutral Certified|Carbon_Neutral_Certified|Carbon-Neutral-Certified|'\
           r'Cruelty-free|Cruelty_free|Cruelty free|'\
           r'Paraben-free|Paraben_free|Paraben free|Natural ingredients|Natural-ingredients|Natural_ingredients|linen|cotton|vegan leather|vegan_leather|vegan-leather'
```

```python
categories = categories.rename(columns = {'id' : 'category_id'})
categories
```

```python
amazon['green'] = amazon['title'].str.contains(keywords, case=False, regex=True)
amazon = pd.merge(amazon, categories, on = 'category_id' , how = 'left')
amazon
```

```python
#amazon = amazon.drop(columns = 'green_cat')

green_categories = ["Girls' Clothing", "Men's Clothing", "Boys' Clothing",
       'Craft & Hobby Fabric', "Baby Boys' Clothing & Shoes",
       "Women's Handbags", 'Bedding', 'Knitting & Crochet Supplies',
       'Skin Care Products', 'Needlework Supplies', 'Baby',
       "Women's Clothing", "Men's Accessories", 'Bath Products',
       'Nursery Furniture, Bedding & Décor', 'Baby Care Products',
       'Hair Care Products', 'Household Cleaning Supplies',
       "Women's Jewelry", "Boys' Jewelry"]
def green_function (category):
  if category in green_categories:
    return True
  else:
    return False

amazon['green_cat'] = amazon['category_name'].apply(green_function)
amazon
```

```python
amazon_green_categories = amazon[amazon['green_cat'] == True]
amazon_green_categories['category_name'].nunique()
```
